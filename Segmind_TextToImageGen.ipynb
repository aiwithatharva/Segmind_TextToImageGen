```python
from diffusers import StableDiffusionXLPipeline
import torch
pipe = StableDiffusionXLPipeline.from_pretrained("segmind/SSD-1B", torch_dtype=torch.float16, use_safetensors=True, variant="fp16")
pipe.to("cuda")
# if using torch < 2.0
# pipe.enable_xformers_memory_efficient_attention()
```

```python
prompt = "Imagine being a research analyst at a trading firm tasked with gauging the general sentiment about the market in the public. SGD suggests approaching this task by randomly selecting one person on the street, asking their opinion, and publishing the results. " # Your prompt here
neg_prompt = "ugly, blurry, poor quality, grayscale" # Negative prompt here
image = pipe(prompt=prompt, negative_prompt=neg_prompt).images[0]
image
```
